Выполнить пример «Обучение с учителем: классификация набора данных Iris». 

```python
import pandas as pd
iris = pd.read_csv('iris.csv')
```

```python
iris.columns
```

```python
from sklearn.model_selection import train_test_split
X = iris.drop(["species"],axis=1)
y = iris["species"]

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=1)

from sklearn.naive_bayes import GaussianNB 
model = GaussianNB() 
model.fit(Xtrain, ytrain)
y_model = model.predict(Xtest) 

from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

Проверить работу алгоритма при доле тестовой выборки 10% 

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.1,random_state=1)

from sklearn.naive_bayes import GaussianNB 
model = GaussianNB() 

model.fit(Xtrain, ytrain) 
y_model = model.predict(Xtest) 

from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

Проверить работу алгоритма при доле тестовой выборки 50%

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.5, random_state=1)

from sklearn.naive_bayes import GaussianNB 
model = GaussianNB() 

model.fit(Xtrain, ytrain) 
y_model = model.predict(Xtest) 

from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

Выполнить пример «Обучение без учителя: понижение размерности набора данных Iris».

```python
from sklearn.decomposition import PCA # 1. Выбираем класс модели
model = PCA(n_components=2) # 2. Создаем экземпляр модели
# с гиперпараметрами
model.fit(X) # 3. Обучаем модель на данных. Обратите
# внимание, что y мы не указываем!
X_2D = model.transform(X)
```

```python
iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]
sns.lmplot("PCA1", "PCA2", hue='species', data=iris, fit_reg=False);
```

Выполнить пример «Обучение без учителя: кластеризация набора данных Iris»

```python
from sklearn import mixture
model = mixture.GaussianMixture(n_components=3, covariance_type='full') 
model.fit(X) 
y_gmm = model.predict(X)
```

```python
iris['cluster'] = y_gmm
sns.lmplot("PCA1", "PCA2", data=iris, hue='species',
col='cluster', fit_reg=False);
```

Так будут разделены данные, если мы укажем только 2 кластера

```python
from sklearn import mixture
model = mixture.GaussianMixture(n_components=2, covariance_type='full') 
model.fit(X) 
y_gmm = model.predict(X)
```

```python
iris['cluster'] = y_gmm
sns.lmplot("PCA1", "PCA2", data=iris, hue='species',
col='cluster', fit_reg=False);
```

Выполнить пример «Классификация цифр»

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.images.shape
```

```python
X = digits.data
y = digits.target
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(Xtrain, ytrain)
y_model = model.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(ytest, y_model)
sns.heatmap(mat, square=True, annot=True, cbar=False)
plt.xlabel('predicted value') 
plt.ylabel('true value'); 
```

```python
fig, axes = plt.subplots(10, 10, figsize=(8, 8),
subplot_kw={'xticks':[], 'yticks':[]},
gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='binary',interpolation='nearest')
    ax.text(0.05, 0.05, str(y_model[i]), transform=ax.transAxes, color='green' if (ytest[i] == y_model[i]) else 'red')
```
