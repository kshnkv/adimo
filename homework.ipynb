```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.axes as mpl_axes
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


df=pd.read_csv('test.csv',encoding='cp1251')
df
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.hist(df['Age'],bins=7,color='orange')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.boxplot(df['Age']) #в оригинале тут опечатка, просто "х", вместо "ах"
plt.show()
```

```python
sns.violinplot(df['Age'],df['Gender'])
sns.despine()
```

```python
var=df.groupby('Gender').Sales.sum()
fig=plt.figure()
ax1=fig.add_subplot(1,1,1)
var.plot(kind='bar',color='purple')
```

```python
var=df.groupby('BMI').Sales.sum()
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
var.plot(kind='line')
```

```python
var=df.groupby(['BMI','Gender']).Sales.sum()
var.unstack().plot(kind='bar', stacked=True, color=['yellow','blue'],grid=False)
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.scatter(df['Age'],df['Sales'],color='grey')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.scatter(df['Age'],df['Sales'],s=df['Income'],color='green')
plt.show()
```

```python
var=df.groupby(['Gender']).sum().stack()
temp=var.unstack()
type(temp)
x_list=temp['Sales']
label_list=temp.index
plt.pie(x_list,labels=label_list,autopct="%1.1f%%")
plt.show()
```
